### <a id='toc1_1_9_'></a>[Segunda SImulación](#toc0_)

```r
a=create_var(3)
```

```r
datos21=draw.correlated.binary(37, ncol(a), rep(0.5,ncol(a)), abs(a))
n11=10
snps21=draw.correlated.binary(37, n11, rep(0.5,n11), diag(x=1, ncol = n11, nrow = n11))
```

```r
x_continua1=random_norm(datos21[,1])
y_continua1=random_norm(datos21[,1])
```

```r
Y=cbind(y_continua1, datos21[,1])
X=x_continua1
G=cbind(datos21[,2:ncol(datos21)],snps21)
```

```r
P=diag(length(X))-X%*%solve(t(X)%*%X)%*%t(X)
raiz=NULL

for (i in 1:ncol(G)) {
  raiz[i]=sqrt(t(G[,i])%*%P%*%G[,i])
}
```

```r
G_gorro=matrix(ncol = ncol(G),nrow = nrow(G))
for (i in 1:ncol(G)) {
  G_gorro[,i]=(P%*%G[,i])/raiz[i]
}
```

```r
Beta=matrix(ncol=ncol(G), nrow = ncol(Y))

for (i in 1:ncol(Y)) {
  Beta[i,]=t(Y[,i])%*%G_gorro
}
```

```r
sigma2=matrix(ncol=ncol(G), nrow = ncol(Y))

for (i in 1:ncol(G)) {
  for (j  in 1:ncol(Y)) {
    sigma2[j,i]=norm((P%*%Y[,j]-G_gorro[,i]*(Beta[j,i])))^2/(nrow(G)-3)
  }
}
```

```r
Z=Beta/sigma2

##Matríz de covarianzas 
matcov=matrix(ncol = nrow(Z), nrow = nrow(Z))
for (i in 1:nrow(Z)) {
  for (j in 1:nrow(Z)) {
    matcov[i,j]=(t(G[,i])%*%G[,j])/sqrt((t(G[,i])%*%G[,i])*(t(G[,j])%*%G[,j]))
  }
}

matcov
```

```r
valprop=eigen(matcov)$values[1]
vecprop=eigen(matcov)$vectors[,1]
i=1
ET=NULL
for (i in 1:ncol(Z)){
  ET[i]=((t(Z[,i])%*%vecprop)^2)/valprop
}
ET_Pval=pchisq(ET, df = 1)
ET_Pval
```

```r
OT=NULL
for (i in 1:ncol(Z)) {
  OT[i]=t(Z[,i])%*%solve(matcov)%*%(Z[,i])
}

OT_Pval=pchisq(OT, df=6)
```

```r
AT=(1-0.6)*(OT-ET)+ET
AT_Pval=pchisq(AT, df=5)
```

```r
result=tibble(snp=1:ncol(G),ET_Pval,OT_Pval,AT_Pval)
```
